# Django Lets Ride

#### Handy URLS

- http://127.0.0.1:8000/register/ --> Registration of New User
- http://127.0.0.1:8000/registration/login/ --> Login for User
- http://127.0.0.1:8000/logout/ --> Logout User
- http://127.0.0.1:8000/api/create-requesters-request --> API 1 to create a Requesters Request
- http://127.0.0.1:8000/api/create-rider-request --> API 2 to create a Rider Request
- http://127.0.0.1:8000/api/get-requesters-request --> API 3 to list all the requests of Requesters 
- http://127.0.0.1:8000/api/get-rider-request --> API 4 to list all the matching Rider requests for Requesters Request
- http://127.0.0.1:8000/api/update-rider-request --> API to update the applied field of Rider request
